```python
'''Calcular histograma médio''' 
from skimage import io
import cv2
bins = 4096
hist_sum = np.zeros(bins)
for i in range(len(archives_PL_list)):
    SHG_image = ski.io.imread(os.path.join(path, correspondent_archives_SHG[i]))[:, :, 1]
    SHG_image_normalizada = SHG_image/correspondent_accumulation_SHG[i]
    blur = cv2.GaussianBlur(SHG_image_normalizada, (5,5), 0)
    ruido_estimado = abs(cv2.subtract(SHG_image_normalizada, blur)).astype(np.uint32)
    hist, _ = np.histogram(ruido_estimado, bins=bins, range=(0, 4095))
    hist_sum += hist
hist_mean = hist_sum / len(archives_PL_list)
plt.plot(hist_mean)
plt.title("Histograma Médio (0–4095)")
plt.xlabel("Intensidade")
plt.ylabel("Frequência média")
plt.xlim(0, 200)   # <-- limita o eixo horizontal
plt.show()
```

```python
'''Ajuste de curva sobre histograma'''
# Valores de intensidade (0 a 4095)
valores = list(range(4096))

# Contagens médias do histograma
contagens = hist_mean.tolist()

#Índice do primeiro valor de intensidade cuja contagem é nula
indice_x_max = np.where(np.array(contagens) == min(contagens))[0][0] 

# Intensidade com contagem máxima
intensidade_contagem_max = valores[np.argmax(hist_mean)]

# Criar lado adjascente da curva suposta Gaussiana

# Posição em torno da qual o gráfico será refletido
x0 = intensidade_contagem_max

# Convertendo listas para arrays 
valores = np.array(valores[:indice_x_max])
contagens = np.array(contagens[:indice_x_max])

# Refletindo os valores em torno de x0
valores_refletidos = 2 * x0 - valores

# Mantendo as mesmas contagens
contagens_refletidas = contagens.copy()


# À esquerda de x = 3 deve prevalescera curva refletida. À direita, a original
valores_truncados = valores[np.where(valores == x0)[0][0]:]
contagens_truncadas = contagens[np.where(valores == x0)[0][0]:]
valores_refletidos_truncados = valores_refletidos[np.where(valores_refletidos == x0)[0][0]+1:]
contagens_refletidas_truncadas = contagens_refletidas[np.where(valores_refletidos == x0)[0][0]+1:]

# Concatena os respectivos pares de listas sem que o ponto em que elas se encontram seja repetido
valores_totais = list(valores_refletidos_truncados[::-1]) + list(valores_truncados)
contagens_totais = list(contagens_refletidas_truncadas[::-1]) + list(contagens_truncadas)

# Plotando
#plt.plot(valores_truncados, contagens_truncadas, label="Original")
#plt.plot(valores_refletidos_truncados, contagens_refletidas_truncadas, label="Refletida")

plt.plot(valores, contagens, label="Original")
plt.plot(valores_refletidos, contagens_refletidas, label="Refletida")
plt.plot(valores_totais, contagens_totais, label = 'União das curvas', color = 'green', linestyle = ':')
plt.xlabel("Intensidade")
plt.ylabel("Frequência média")
plt.xlim(-100, 100)
plt.axvline(x=intensidade_contagem_max, color='red', linestyle='--', label=f"x = {intensidade_contagem_max}")
plt.legend()
plt.show()

#Normalizando contagens
contagens_totais = list(np.array(contagens_totais)/sum(contagens_totais))

#Ajuste Gaussiano
from scipy.optimize import curve_fit

def gaussian(x, a, mu, sigma):
    """
    Função Gaussiana para ajuste de curva.

    Parâmetros:
    ----------
    x: array
        Variável independente.
    a: float
        Amplitude máxima da curva.
    mu: float
        Média da curva (centro).
    sigma: float
        Desvio padrão da curva (largura).
    """
    return a * np.exp(-(x - mu)**2 / (2 * sigma**2))


popt, pcov = curve_fit(gaussian, valores_totais, contagens_totais, p0=[max(contagens_totais), 
                                                                       intensidade_contagem_max, 
                                                                       np.std(np.array(contagens_totais))])
print(popt)
'''
popt: array com os parâmetros ótimos ajustados (a, mu, sigma)
pcov: matriz de covariância dos parâmetros
gaussian: função definida, que pretende-se ajustar à dispersão dos dados
p0: chutes iniciais para os parâmetros

'''

# Gera a curva ajustada usando os parâmetros otimizados
y_ajustado = gaussian(valores_totais, *popt)

plt.figure(figsize=(8, 6))
plt.plot(valores_totais, y_ajustado, label='Curva Ajustada', color='red')
plt.scatter(valores_totais, contagens_totais, label='Dados Reais', color='blue')
plt.title('Ajuste de Curva Gaussiana')
plt.xlabel('Intensidade')
plt.ylabel('Contagens Normalizadas')
plt.legend()
plt.grid(True)
plt.xlim(-50,50)
plt.show()
```

# Curva de Cauchy-Lorentz
$f(x)=\frac{A}{1+\left(\frac{x-x_0}{\gamma}\right)^2}$, ou, em sua forma normalizada, $f(x)=\frac{1}{\pi\gamma\left[1+\left(\frac{x-x_0}{\gamma}\right)^2\right]}$, onde $A$ é a amplitude, $x_0$ é o a posição do pico central e $\gamma$ a largura da curva a meia altura.



```python
from numpy import interp
def estimar_pars(x,y):
    y_max = max(y)
    x_0 = x[np.argmax(y)]
    
    #Converte listas para arrays
    x = np.array(x)
    y = np.array(y)
    
    # Separa lados direito e esquerdo da curva
    x_left = x[x < x_0]
    y_left = y[x < x_0]
    x_right = x[x > x_0]
    y_right = y[x > x_0]
    
    #Interpola para achar os valores aproximados da coordenada x para os pontos a meia altura
    x_left_half = interp(y_max/2, y_left[::-1], x_left[::-1])  # do lado esquerdo
    x_right_half = interp(y_max/2, y_right, x_right)           # do lado direito
    
    #Largura aproximada a meia altura
    fwhm_guess = (x_right_half - x_left_half)/2
    
    return y_max, x_0, fwhm_guess

def cauchy_lorentz(x, A, x_0, gamma):
    """
    Função de Cauchy-Lorentz para ajuste de curva.

    Parâmetros:
    ----------
    x: array
        Variável independente.
    A: float
        Amplitude máxima da curva.
    x_0: float
        Posição do pico central.
    gamma: float
        Largura a meia altura.
    """
    return A/(1+((x-x_0)/(gamma))**2)


popt_cl, pcov_cl = curve_fit(cauchy_lorentz, valores_totais, contagens_totais, p0=[estimar_pars
                                                                                   (valores_totais, 
                                                                                    contagens_totais)])
#Plota dispersão dos dados e ajuste

a, x0, Gamma = popt_cl
eq_text = fr'$f(x)=\frac{{{a:.3f}}}{{1+\left(\frac{{x-{x0:.3f}}}{{{Gamma:.3f}}}\right)^2}}$'

y_ajustado_cauchy_lorentz = cauchy_lorentz(valores_totais, *popt_cl)

plt.figure(figsize=(8, 6))
plt.plot(valores_totais, y_ajustado_cauchy_lorentz, label='Curva Ajustada', color='red')
plt.scatter(valores_totais, contagens_totais, label='Dados Reais', color='blue')
plt.title('Ajuste de Curva Cauchy-Lorentz')
plt.xlabel('Intensidade')
plt.ylabel('Contagens Normalizadas')

plt.text(0.05, 0.95, eq_text, transform=plt.gca().transAxes,fontsize=15, 
         verticalalignment='top', bbox=dict(boxstyle="round,pad=0.5", 
                                            facecolor="white", alpha=0.7))
plt.legend()
plt.grid(True)
plt.xlim(-50,50)
plt.show()
```

Evidentemente, não tenho interesse nos valores negativos. A ideia de espelhar a curva
em relação à coordenada horizontal correspondente ao pico foi estritamente para facilitar o ajuste de curva.
Quero usar a equação deduzida para gerar ruídos pseudoaleatórios, segundo essa distribuição. A função
de probabilidade $p(x)$ estará definida para $0\le x\le M$, onde M é o valor máximo de intensidade que se verifica no ruído; logo, a área sob o gráfico nesse intervalo deve ser unitária. Calcula-se:

$\displaystyle\sum_{x=0}^{M}\frac{A}{1+\left(\frac{x-x_0}{\gamma}\right)^2}$. Basta dividir a expressão anterior, obtida pelo ajuste e evidente no gráfico pelo resultado do somatório, para obter a função de probabilidade.

```python
import sympy as sp
from sympy import symbols, latex
from IPython.display import display, Math

# Define as variáveis simbólicas e as funções
A, gamma, M, x_0, x= symbols('A gamma M x_0 x')
termo = A / (1 + ((x - x_0) / gamma)**2)
somatoria = sp.Sum(termo, (x, 0, M))

eq_A = fr'A = {a:.4f}'
eq_x_0 = f'x_0 = {x0:.0f}'
eq_gamma = fr'$\gamma = {Gamma:.4f}'
eq_M = f'M = {max(valores_totais)}'

# Converta a função para uma string LaTeX
latex_string = latex(somatoria)

# Exibe as strings como equações renderizadas
display(Math(eq_A))
display(Math(eq_x_0))
display(Math(eq_gamma))
display(Math(eq_M))

# Formata o somatório com as substituições para printar ao final 

termo_numerico_str = fr'\frac{{{a:.4f}}}{{1+\left(\frac{{x-{x0:.0f}}}{{{Gamma:.4f}}}\right)^2}}'
somatoria_numeros_str = fr'\sum_{{x=0}}^{{{max(valores_totais)}}} {termo_numerico_str}'

# Cria a lista de valores para o somatório
valores_somatorio = np.array(range(max(valores_totais)+1))

# Cria uma função numérica com o módulo lambdify do sympy que pode receber arrays de entrada, 
# o que é computacionalmente eficiente para calcular todos os termos do somatório de maneira vetorizada.

funcao_numerica = sp. lambdify([x, A, x_0, gamma], termo, 'numpy')

valores_dos_termos = funcao_numerica(valores_somatorio, a, x0, Gamma)

# Calcula o valor do somatório
soma_numerica = np.sum(valores_dos_termos)


display(Math(fr'{latex(somatoria)} \approx {somatoria_numeros_str} \approx {soma_numerica:.3f}' ))
razao = a/soma_numerica
expressao_final = fr'\frac{{{razao:.5f}}}{{1+\left(\frac{{x-{x0:.0f}}}{{{Gamma:.4f}}}\right)^2}}'
display(Math(fr'p(x)=\frac{{1}}{{{soma_numerica:.3f}}}{termo_numerico_str} \approx {expressao_final}, 0\le x\le{max(valores_totais)}'))

#Função normalizada estimada
def f_continuous(x):
    """densidade contínua dada na imagem"""
    return razao / (1.0 + ((x - a)/Gamma)**2)
  
# domínio discreto
xs = np.arange(0, 596)   # 0..595 inclusive

y_estimado = f_continuous(xs)

plt.figure(figsize=(8, 6))
plt.plot(xs, y_estimado, label='Modelo', color='red')
plt.scatter(valores, contagens/sum(contagens), label='Dados Reais', color='blue')
plt.title('Comparação entre dados experimentais e performance do modelo')
plt.xlabel('Intensidade')
plt.ylabel('Contagens Normalizadas')
plt.legend()
plt.grid(True)
plt.xlim(-50,50)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# parâmetros da tua função
x0 = 3.0
scale = 7.6913
A = 0.06549

def f_continuous(x):
    """densidade contínua dada na imagem"""
    return A / (1.0 + ((x - x0)/scale)**2)

# domínio discreto
xs = np.arange(0, 596)   # 0..595 inclusive

# avaliar a "densidade" em cada inteiro
vals = f_continuous(xs)

# assegurar não-negatividade (por garantia)
vals = np.clip(vals, 0, None)

# normalizar para obter probabilidades discretas que somam 1
prob = vals / vals.sum()

# verificar soma == 1 (aprox.)
print("Soma das probabilidades:", prob.sum())

# gerar amostras pseudo-aleatórias
rng = np.random.default_rng(seed=12345)  # muda/retira seed se não deseja determinismo
N = 20000
samples = rng.choice(xs, size=N, p=prob)

# histograma comparativo (frequências normalizadas vs. distribuição alvo)
plt.figure(figsize=(9,4))

# histograma das amostras (probabilidade empírica)
counts, bins, _ = plt.hist(samples, bins=range(0, 597), density=True, alpha=0.6, label="empírico (amostras)")

# plot da probabilidade teórica discreta
plt.plot(xs, prob, lw=2, label="probabilidade discreta (teórica)")
plt.xlim(0, 595)
plt.xlabel("x (inteiro)")
plt.ylabel("probabilidade")
plt.legend()
plt.tight_layout()
plt.show()
```
